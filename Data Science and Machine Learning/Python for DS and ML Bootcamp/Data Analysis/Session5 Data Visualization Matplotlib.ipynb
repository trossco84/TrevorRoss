```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
#functional
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
##Object Oriented Method
fig = plt.figure()


axes = fig.add_axes([0.1,0.1,0.8,0.8])
#axes are left bottom width and height

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig2 = plt.figure()

axes1=fig2.add_axes([0.1,0.1,0.8,0.8])
axes2=fig2.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes2.plot(y,x)

axes1.set_title('LargePlot')
axes2.set_title('SmallPlot')
```

```python
fig3,axes = plt.subplots(nrows = 1, ncols=2)
#for current_ax in axes:
#    current_ax.plot(x,y)
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
##Figure size and dpi
fig5 = plt.figure(figsize=(8,2),dpi =100)

ax = fig5.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout
```

```python
#saving
fig.savefig('my_firstfig.png',dpi=200)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='x squared')
ax.plot(x,x**3, label = 'x cubed')

#adding a legend
ax.legend(loc=(0.1,0.1))
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
#ax.plot(x,y,color = 'purple',linewidth=2,linestyle='-', marker = 'o', markersize = 10,
      # markerfacecolor = 'yellow',markeredgewidth=3,markeredgecolor='green') 
#RGB Hex Code for colors, linewidth default is 1 also knowns as lw, alpha is transparency,linestyle = ls
#markers specify the points of values, markersize
#look at documentation for documentation parameters matplotlib concepts lecture
ax.plot(x,y,color = 'purple',lw=2,ls='--')
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
